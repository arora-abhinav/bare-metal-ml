```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/abhinavarora/Desktop/Machine Learning/Logistic Regression/Iris.csv")

# Drop the Id column if it exists
if 'Id' in df.columns:
    df = df.drop(columns=['Id'])

df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

test_size = 0.2
split_idx = int(len(df_shuffled) * (1 - test_size))

train_df = df_shuffled.iloc[:split_idx]
test_df = df_shuffled.iloc[split_idx:]

X_train = train_df.iloc[:, :-1].values  # Now correctly 4 feature columns
y_train = train_df.iloc[:, -1].values

X_test = test_df.iloc[:, :-1].values
y_test = test_df.iloc[:, -1].values     # ✅ fixed typo: Y_test -> y_test for consistency

print(f"X_train shape: {X_train.shape}")  # Should be (120, 4)
print(f"X_test shape: {X_test.shape}")    # Should be (30, 4)
```

```python
def activation(params, features):
    sum = params[0]
    for i in range(len(features)):
        sum += params[i + 1] * features[i]
    
    if sum < 0:
        return 0
    return 1
```

```python
def learn(initial_params, X_TRAIN, iterations, Y_TRAIN, learning_rate):
    final_params = initial_params.copy()
    for _ in range(iterations):
        thetas = np.zeros(len(initial_params))
        for i in range(len(X_TRAIN)):
            a = activation(final_params, X_TRAIN[i])
            y = 1
            if Y_TRAIN[i] == 'Iris-versicolor':
                y = 0
            error = y - a

            for k in range(len(thetas)):
                if k == 0:
                    thetas[k] += error

                else:
                    thetas[k] += error * X_TRAIN[i][k - 1]
        
            for l in range(len(final_params)):
                final_params[l] += learning_rate * thetas[l]
    
    return final_params
```

```python
def predict(params, X):
    predictions = []
    for features in X:
        a = activation(params, features)
        if a == 1:
            predictions.append('Iris-versicolor')
        else:
            predictions.append('Iris-non-versicolor')
    return predictions


# Train
trained_params = learn(np.zeros(5), X_train, 1000, y_train, 0)

# Predict
y_pred = predict(trained_params, X_test)

correct = 0
for pred, actual in zip(y_pred, y_test):
    actual_binary = 1 if actual == "Iris-versicolor" else 0
    pred_binary = 1 if pred == "Iris-versicolor" else 0
    if pred_binary == actual_binary:
        correct += 1

accuracy = correct / len(y_test) * 100
print(accuracy)


```
