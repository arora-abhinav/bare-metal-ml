```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("score.csv")
data.head(100)
entries = data.shape[0]
print(data)
```

```python
#First: Training Data is required
x = data["Hours"]
y = data["Scores"]
plt.scatter(x, y)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Hours Studied vs Score")
plt.show()
```

```python
#Creating the loss function:
def loss_function(theta0, theta1):
    sum = 0
    for i in range(entries):
        sum += 0.5 * ((theta0 + theta1 * data.iloc[i]["Hours"]) - (data.iloc[i]["Scores"])) ** 2

def hypothesis(theta0, theta1, weight):
    return theta0 + theta1 * weight

def gradient_descent(learning_rate, initial_theta0, initial_theta1, iterations):
    theta0 = initial_theta0
    theta1 = initial_theta1
    entries = len(x)
    
    for _ in range(iterations):
        # Calculate gradients
        sum_error_theta0 = 0
        sum_error_theta1 = 0
        
        for i in range(entries):
            hours = x.iloc[i]
            score = y.iloc[i]
            h = hypothesis(theta0, theta1, hours)
            error = h - score
            sum_error_theta0 += error
            sum_error_theta1 += error * hours
        
        # Update parameters (gradient descent step)
        theta0 -= (learning_rate / entries) * sum_error_theta0
        theta1 -= (learning_rate / entries) * sum_error_theta1
    
    return [theta0, theta1]

def plot():
    feature = np.linspace(0, 10, 1000)
    theta = gradient_descent(0.01, 0, 0, 100)
    print(theta)
    
    # Plot original data points
    plt.scatter(x, y, label='Data')
    
    # Plot fitted line
    output = theta[0] + feature * theta[1]
    plt.plot(feature, output, 'r-', label='Fitted line')
    
    plt.xlabel('Hours')
    plt.ylabel('Score')
    plt.legend()
    plt.show()

plot()


```
