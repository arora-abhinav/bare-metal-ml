```python
import pandas as p
import numpy as np
import math
```

```python
df = p.read_csv('/Users/abhinavarora/Desktop/Machine Learning/Logistic Regression/Iris.csv')
print(df)
```

```python
sepal_length = df["SepalLengthCm"].to_numpy()
sepal_width = df["SepalWidthCm"].to_numpy()
petal_length = df["PetalLengthCm"].to_numpy()
petal_width = df["PetalWidthCm"].to_numpy()
species = df["Species"].to_numpy()
print(species)
```

```python
def hypothesis(params, features):
    lin_comb = params[0]
    for i in range(len(features)):
        lin_comb += params[i + 1] * features[i]
    res = 1/(1 + math.exp(-1 * lin_comb))
    return res
```

```python
def batch_gradient_ascent(learning_rate, initial_params, features, iterations, training_length):
    final_params = initial_params.copy()
    for _ in range(iterations):
        thetas = initial_params.copy()
        for i in range(training_length):
            f = []
            for j in features:
                f.append(j[i])
            h = hypothesis(final_params, f)
            flower = species[i]
            y = 0
            if flower == 'Iris-setosa':
                y = 1
            
            error = y - h
            for k in range(len(thetas)):
                if k == 0:
                    thetas[k] += error
                else:
                    thetas[k] += error * f[k - 1]
        
        for l in range(len(final_params)):
            final_params[l] += (learning_rate * thetas[l])/training_length
    
    return final_params
            
            
```

```python

final_thetas = batch_gradient_ascent(0.001, np.zeros(5), [sepal_length[:100], sepal_width[:100], petal_length[:100], petal_width[:100]], 1000, 100)
print(final_thetas)

def create_predictions():
    # Make predictions
    predictions = []
    for i in range(len(sepal_length)):
        # Create feature vector for example i
        f = [sepal_length[i], sepal_width[i], petal_length[i], petal_width[i]]
        
        # Get probability
        h = hypothesis(final_thetas, f)
        
        # Classify based on threshold
        if h >= 0.5:
            predictions.append(1)  # Predict Iris-setosa
        else:
            predictions.append(0)  # Predict not Iris-setosa

    # Compare with actual labels
    actual = [1 if s == "Iris-setosa" else 0 for s in species]

    # Calculate accuracy
    correct = sum([1 for i in range(len(predictions)) if predictions[i] == actual[i]])
    accuracy = correct / len(predictions)
    print(f"Accuracy: {accuracy * 100:.2f}%")

    # Show some predictions
    print("\nFirst 100 predictions:")
    for i in range(100):
        print(f"Example {i}: Predicted={predictions[i]}, Actual={actual[i]}, Probability={hypothesis(final_thetas, [sepal_length[i], sepal_width[i], petal_length[i], petal_width[i]]):.3f}")

create_predictions()
```
